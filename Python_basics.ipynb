```python
## Lets plot

# Importing matplotlib library
import matplotlib.pyplot as plt
plt.figure(num= '1', figsize=(8,8)) # figure size and number
plt.plot(x,y**2 +2, color='blue') # line plot of x and y^2 +2
plt.scatter(x,y**2 +2, color='orange') # scatter point of x and y^2 +2
plt.xlabel('X')
plt.ylabel('Y**2 +2')
plt.title('X vs Y')
#plt.savefig('XvsY.png') # save figure
plt.show()
```

```python
lon=np.arange(0,360+1,1)
lat=np.arange(-90,90+1, 1)

# Don't need to understand this part. a function genrating data
Tadd=np.zeros((len(lat),len(lon)))
from math import exp
for i in range(len(lon)):
    x = exp(-1*((lon[i]-150)**2)/(50**2))-exp(-1*((lon[i]-210)**2)/(50**2))
    for j in range(len(lat)):
        y = exp(-1*((lat[j]+5)**2)/(10**2))
        Tadd[j,i]= 86500*x*y
        
#### ploting the data Tadd
plt.figure(num= '1', figsize=(11, 5.5)) # figure size and number
plt.contourf(lon,lat,Tadd,cmap='bwr')
plt.colorbar()
plt.xlabel('Lon')
plt.ylabel('Lat')
plt.title('Generated data')
#plt.savefig('XvsY.png' )
plt.show()
```

```python
#### ploting the data Tadd
import aoespy as ap # Loading aoespy to plot with map
plt.figure(num= '1', figsize=(11, 5.5)) # figure size and number
ap.plotc(Tadd,lon,lat,cmap='bwr')
plt.title('Generated data')
#plt.savefig('XvsY.png' )
plt.show()
```
